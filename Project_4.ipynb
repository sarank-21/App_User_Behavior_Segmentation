## **K-Means**

```python
app_df_encode = app_df.copy()
```

```python
app_df_encode.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Fix 1: Use only the 3 strongest behavioral signals
best_cols = ['sessions_per_week', 'avg_session_duration_min',
       'daily_active_minutes', 'feature_clicks_per_session',
       'notifications_opened_per_week', 'in_app_search_count',
       'pages_viewed_per_session', 'crash_events_last_30_days',
       'support_tickets_raised', 'days_since_last_login',
       'ads_clicked_last_30_days', 'content_downloads', 'social_shares',
       'rating_given', 'churn_risk_score', 'engagement_score',
       'account_age_days']

X = app_df[best_cols].copy()

# Fix 2: Scale first
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fix 3: Apply PCA to compress and remove noise
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# print(f"Variance retained: {sum(pca.explained_variance_ratio_)*100:.1f}%")

# Try different k values on the PCA output
km = KMeans(n_clusters=2,
            random_state=42,
            n_init=20,
            max_iter=500
            )
labels = km.fit_predict(X_scaled)
score = silhouette_score(X_scaled, labels)
print(f"k={2}, Silhouette Score={score:.4f}")

app_df['cluster'] = km.fit_predict(X_scaled)

app_df['user_segment'] = app_df['cluster'].map({
    0: 'Casual Users',
    1: 'Power Users'})
```

```python
app_df['user_segment'].value_counts()
```

```python
# ELBOW METHOD — Finding the right number of clusters
# WCSS = how spread out points are within each cluster (lower = tighter clusters)
import matplotlib.pyplot as plt
wcss = []
K_values = range(1, 10)

for k in K_values:
    model = KMeans(n_clusters=k,
            random_state=42,
            n_init=20,
            max_iter=500
            )
    model.fit(X_scaled)
    wcss.append(model.inertia_)   # inertia = WCSS

# Plot the elbow
plt.figure(figsize=(10, 5))
plt.plot(K_values, wcss, 'o-', color='#3498db', linewidth=3, markersize=10, markerfacecolor='white',
         markeredgewidth=2, markeredgecolor='#3498db')



plt.title('🦾 Elbow Method — How Many Clusters?', fontsize=14, fontweight='bold')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Total cluster tightness)')
plt.xticks(K_values)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.4)
plt.tight_layout()
plt.show()
```

# **DBSCAN**

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Fix 1: Use only the 3 strongest behavioral signals
best_cols = ['sessions_per_week', 'avg_session_duration_min',
       'daily_active_minutes', 'feature_clicks_per_session',
       'notifications_opened_per_week', 'in_app_search_count',
       'pages_viewed_per_session', 'crash_events_last_30_days',
       'support_tickets_raised', 'days_since_last_login',
       'ads_clicked_last_30_days', 'content_downloads', 'social_shares',
       'rating_given', 'churn_risk_score', 'engagement_score',
       'account_age_days']

X = app_df[best_cols].copy()

# Fix 2: Scale first
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# PCA (keep 90% variance)
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X)
```

```python
# --- DBSCAN ---
dbscan = DBSCAN(eps=0.99, min_samples=10)
labels_dbscan = dbscan.fit_predict(X_pca)

# Count results
n_clusters = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
n_noise    = list(labels_dbscan).count(-1)

print(f'DBSCAN Results:')
print(f'  Clusters found:  {n_clusters}')
print(f'  Noise points:    {n_noise} flowers flagged as outliers')
unique, counts = np.unique(labels_dbscan, return_counts=True)
for u, c in zip(unique, counts):
    label = 'NOISE ⚫' if u == -1 else f'Cluster {u}'
    print(f'  {label}: {c} points')
```

```python
from sklearn.metrics import silhouette_score

# Ignore noise points (-1)
mask = labels_dbscan != -1

if len(set(labels_dbscan[mask])) > 1:
    score = silhouette_score(X_pca[mask], labels_dbscan[mask])
    print(f"Silhouette Score: {score:.4f}")
else:
    print("Not enough clusters for Silhouette Score")
```
